```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("clean_dataxx.csv")
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps = 10, min_samples = 10)
```

```python
data.columns
```

```python
numeric_data = data[['CS', 'Biology', 'Chemistry', 'Math', 'Further Math',
       'Physics', 'Ofsted_grade', 'student_language']]
```

```python
encoding = pd.get_dummies(numeric_data.Ofsted_grade, drop_first= True)
encoded_data = pd.concat([numeric_data, encoding], axis = 1)
encoded_data = encoded_data.drop(['Ofsted_grade'], axis = 1)
encoded_data
```

```python
encoded_data  = encoded_data.dropna()
```

```python
model = dbscan.fit(encoded_data)
```

```python
labels = model.labels_
```

```python
model.labels_
```

```python
np.unique(labels)
```

### SO it clusters the values on 17 clusters

```python
encoded_data["labels"] = labels
encoded_data["lat"] = data['latitude']
encoded_data["lon"] = data['longitude']
```

```python
encoded_data
```

```python
encoded_data.to_csv("dbscan.csv")
```
