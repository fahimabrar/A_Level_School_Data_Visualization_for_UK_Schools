#### converting percentage to total number

```python
def percentage_to_total(x, y):
    return round(x*y/100)
```

```python
df = df[['LAESTAB', 'SCHNAME.x', 'ALL_RUSSELL',
       'ALL_OXBRIDGE', 'A_level_maths_percent_all',
       'A_level_further_maths_percent_all',
       'A_level_biological_sciences_percent_all',
       'A_level_chemistry_percent_all', 'A_level_physics_percent_all',
       'A_level_computer_sciences_percent_all', 'Female_A_levels_students',
       'A_level_maths_percent_female', 'A_level_further_maths_percent_female',
       'A_level_biological_sciences_percent_female',
       'A_level_chemistry_percent_female', 'A_level_physics_percent_female',
       'A_level_computer_sciences_percent_female', 'Male_A_levels_students',
       'A_level_maths_percent_male', 'A_level_further_maths_percent_male', 
       'A_level_biological_sciences_percent_male',
       'A_level_chemistry_percent_male', 'A_level_physics_percent_male',
       'Alevel_computer_sciences_percent_male','Ofsted_grade',
       'latitude', 'longitude', 'NUMEAL', 'NUMUNCFL']]
```

```python
df
```

```python
def remove_per(x):
    x = str(x)
    x = x.strip()
    if x == "nan":
        return 0
    else:
        return x.replace('%', '')
```

```python
df['ALL_RUSSELL'] = df['ALL_RUSSELL'].apply(remove_per)
df['ALL_OXBRIDGE'] = df['ALL_OXBRIDGE'].apply(remove_per)
```

```python
n_data["CS_male"] = round(df['Male_A_levels_students']* df['Alevel_computer_sciences_percent_male'] /100)
n_data["CS_female"] = round(df['Female_A_levels_students']* df['A_level_computer_sciences_percent_female'] /100)
n_data["Biology_male"] = round(df['Male_A_levels_students']* df['A_level_biological_sciences_percent_male'] /100)
n_data["Biology_female"] = round(df['Female_A_levels_students']* df['A_level_biological_sciences_percent_female'] /100)
n_data["Chemistry_male"] = round(df['Male_A_levels_students']* df['A_level_chemistry_percent_male'] /100)
n_data["Chemistry_female"] = round(df['Female_A_levels_students']* df['A_level_chemistry_percent_female'] /100)
n_data["Math_male"] = round(df['Male_A_levels_students']* df['A_level_maths_percent_male'] / 100)
n_data["Math_female"] = round(df['Female_A_levels_students']* df['A_level_maths_percent_female'] / 100)
n_data["Furhter_Math_male"] = round(df['Male_A_levels_students']* df['A_level_further_maths_percent_male'] /100)
n_data["Furhter_Math_female"] = round(df['Female_A_levels_students']* df['A_level_further_maths_percent_female'] /100) 
n_data["Physics_male"] = round(df['Male_A_levels_students']* df['A_level_physics_percent_male'] /100)
n_data["Physics_female"] = round(df['Female_A_levels_students']* df['A_level_physics_percent_female']/100)
```
