```python
def update_name(x):
    try:
        laestab = x
        year =df[df["LAESTAB"] == laestab]["YEAR"].max()
        #finding the most recent year by applying max fuction for year, for a specific laestab
        updated_schl_name = df[(df["LAESTAB"] == laestab) & (df["YEAR"] == year)]["SCHNAME"]
        #getting the school name for this year for this same laestab
        return updated_schl_name.tolist()[0]
        
    except:
        print("skipping")
```

```python
df["SCHNAME"] = df["LAESTAB"].apply(update_name)
```
